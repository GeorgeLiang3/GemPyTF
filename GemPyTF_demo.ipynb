```python
# %%
import sys
sys.path.append('./GemPyTF/GP_old')

import gempy as gp

from gempy import map_series_to_surfaces
from gempy.core.tensor.modeltf_var import ModelTF

# %%
data_path = 'https://raw.githubusercontent.com/cgre-aachen/gempy_data/master/'
path_to_data = data_path + "/data/input_data/jan_models/"
```

## model1-Horizontalstratigraphic

3D plots requires vtk, please use local kernel.

```python

geo_data = gp.create_data( extent=[0, 1000, 0, 1000, 0, 1000], resolution=[50, 50, 50],
                          path_o=path_to_data + "model1_orientations.csv",
                          path_i=path_to_data + "model1_surface_points.csv")


map_series_to_surfaces(geo_data, {"Strat_Series": ('rock2', 'rock1'), "Basement_Series": ('basement')})

# %%
## the module into GemPy through Interpolator later
model = ModelTF(geo_data)
model.activate_regular_grid()
gpinput = model.get_graph_input()
model.create_tensorflow_graph(gpinput,gradient = False)

# %%
model.compute_model()
gp.plot.plot_section(model, cell_number=15,
                         direction='y', show_all_data= True)
```

## model7-combination

```python
geo_data7 = gp.create_data( extent=[0, 2500, 0, 1000, 0, 1000],
                          resolution=[125, 50, 50],
                          path_o=path_to_data + "model7_orientations.csv",
                          path_i=path_to_data + "model7_surface_points.csv")

# %%
gp.map_series_to_surfaces(geo_data7, {"Fault_Series": 'fault', "Strat_Series1": ('rock3'),
                                     "Strat_Series2": ('rock2','rock1')})


geo_data7.set_is_fault(['Fault_Series'])

# %%

model7 = ModelTF(geo_data7)
model7.activate_regular_grid()
gpinput = model7.get_graph_input()
model7.create_tensorflow_graph(gpinput,gradient = False)


# %%
model7.compute_model()

# %%
gp.plot.plot_section(model7, cell_number=15,
                         direction='y', show_data=True,show_all_data = True)
```
